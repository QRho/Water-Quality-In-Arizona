```python
import pandas as pd
import requests
import gmaps
from config import gkey 
```

```python
gmaps.configure(api_key=gkey)
```

```python
api_url = "https://www.waterqualitydata.us/ogcservices/wfs/?request=GetFeature&service=wfs&version=2.0.0&typeNames=wqp_sites&SEARCHPARAMS=countrycode%3AUS%3Bstatecode%3AUS%3A04%3Bcountycode%3AUS%3A04%3A013%3BsiteType%3ALake%2C%20Reservoir%2C%20Impoundment%3BsampleMedia%3Awater%7CWater&outputFormat=application%2Fjson"

response = requests.get(api_url).json()
```

```python
#print(response["features"][0])
coordinates = []
for feature in response["features"]:
    latitude = feature["geometry"]["coordinates"][1]
    longitude = feature["geometry"]["coordinates"][0]
    coordinates.append(tuple((latitude, longitude)))

coordinates
```

```python
figure_layout = {
    'width': '400px',
    'height': '300px',
    'border': '1px solid black',
    'padding': '1px',
    'margin': '0 auto 0 auto'
}
fig = gmaps.figure(layout=figure_layout)  # Assign the marker layer to a variable
markers = gmaps.marker_layer(coordinates)
# Add the layer to the map
fig.add_layer(markers)
fig
```
